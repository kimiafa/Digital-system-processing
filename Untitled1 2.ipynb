<b>3.1.</b> تبدیل Z را همراه با ناحیه همگرایی (ROC) برای هر یک از دنباله‌های زیر تعیین کنید.
</div>

#### (a)
$$
\mathcal{Z} \left[ \left(\frac{1}{2} \right)^n u[n] \right] = \sum_{n=0}^{\infty} \left(\frac{1}{2} \right)^n z^{-n} = \sum_{n=0}^{\infty} \left(\frac{1}{2z} \right)^n = \frac{1}{1 - \frac{1}{2}z^{-1}}, \quad |z| > \frac{1}{2}
$$

#### (b)
$$
\mathcal{Z} \left[ -\left(\frac{1}{2} \right)^n u[-n-1] \right] = -\sum_{n=-\infty}^{-1} \left(\frac{1}{2} \right)^n z^{-n} = -\sum_{n=1}^{\infty} (2z)^n
$$
$$
= \frac{-2z}{1 - 2z} = \frac{1}{1 - \frac{1}{2}z^{-1}}, \quad |z| < \frac{1}{2}
$$

#### (c)
$$
\mathcal{Z} \left[ \left(\frac{1}{2} \right)^n u[-n] \right] = \sum_{n=-\infty}^{0} (2z)^n = \frac{1}{1 - 2z}, \quad |z| < \frac{1}{2}
$$

#### (d)
$$
\mathcal{Z} [\delta[n]] = z^0 = 1, \quad \text{all } z
$$

#### (e)
$$
\mathcal{Z} [\delta[n-1]] = z^{-1}, \quad |z| > 0
$$

#### (f)
$$
\mathcal{Z} [\delta[n+1]] = z^{+1}, \quad 0 \leq |z| < \infty
$$

#### (g)
$$
\mathcal{Z} \left[ \left(\frac{1}{2} \right)^n (u[n] - u[n-10]) \right] = \sum_{n=0}^{9} \left(\frac{1}{2z} \right)^n
$$
$$
= \frac{1 - (2z)^{-10}}{1 - (2z)^{-1}}, \quad |z| > 0
$$

<b>3.2.</b> تبدیل Z دنباله زیر را تعیین کنید.
</div

$$
x[n] =
\begin{cases}
n, & 0 \leq n \leq N-1 \\
N, & N \leq n
\end{cases}
= n u[n] - (n - N) u[n - N]
$$

$$
n \, x[n] \quad \Longleftrightarrow \quad -z \frac{d}{dz} X(z) \Rightarrow n \, u[n] \quad \Longleftrightarrow \quad -z \frac{d}{dz} \frac{1}{1 - z^{-1}}, \quad |z| > 1
$$

$$
n \, u[n] \quad \Longleftrightarrow \quad \frac{z^{-1}}{(1 - z^{-1})^2}, \quad |z| > 1
$$

$$
x[n - n_0] \quad \Longleftrightarrow \quad X(z) \cdot z^{-n_0} \Rightarrow (n - N) u[n - N] \quad \Longleftrightarrow \quad \frac{z^{-N-1}}{(1 - z^{-1})^2}, \quad |z| > 1
$$

<div dir="rtl" align="right">
**بنابراین:**
</div>

$$
X(z) = \frac{z^{-1} - z^{-N-1}}{(1 - z^{-1})^2} = \frac{z^{-1} (1 - z^{-N})}{(1 - z^{-1})^2}
$$

<div dir="rtl" align="right">
<b>3.3.</b> تبدیل Z هر یک از دنباله‌های زیر را تعیین کنید. پاسخ خود را همراه با ناحیه همگرایی (ROC) در صفحه Z و یک نمودار قطب-صفر ارائه دهید. تمامی مجموع‌ها را به صورت فرم بسته بیان کنید؛ مقدار آلفا می‌تواند مختلط باشد.
</div>

$$
\textbf{(a)}
$$

$$
x_a[n] = \alpha^{|n|}
\quad 0 < |\alpha| < 1
$$

$$
X_a(z) = \sum_{n=-\infty}^{-1} \alpha^{-n}z^{-n} + \sum_{n=0}^{\infty} \alpha^n z^{-n}
$$

$$
= \sum_{n=1}^{\infty} \alpha^n z^n + \sum_{n=0}^{\infty} \alpha^n z^{-n}
$$

$$
= \frac{\alpha z}{1 - \alpha z} + \frac{1}{1 - \alpha z^{-1}}
$$

$$
= \frac{z(1 - \alpha^2)}{(1 - \alpha z)(z - \alpha)},
\quad |\alpha| < |z| < \frac{1}{|\alpha|}
$$

---

$$
\textbf{(b)}
$$

$$
x_b[n] =
\begin{cases}
1, & 0 \leq n \leq N-1 \\
0, & N \leq n \\
0, & n < 0
\end{cases}
$$

$$
X_b(z) = \sum_{n=0}^{N-1} z^{-n} = \frac{1 - z^{-N}}{1 - z^{-1}}
$$

$$
= \frac{z^N - 1}{z^{N-1} (z - 1)},
\quad z \neq 0
$$

---

$$
\textbf{(c)}
$$

$$
x_c[n] = x_b[n-1] * x_b[n]
$$

$$
\Rightarrow X_c(z) = z^{-1} X_b(z) \cdot X_b(z)
$$

$$
X_c(z) = z^{-1} \left( \frac{z^N - 1}{z^{N-1} (z - 1)} \right)^2
$$

$$
= \frac{1}{z^{2N-1}} \left( \frac{z^N - 1}{z - 1} \right)^2,
\quad z \neq 0, 1
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_pole_zero(poles, zeros, title):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.scatter(np.real(zeros), np.imag(zeros), s=100, marker='o', facecolors='none', edgecolors='b', label="Zeros")
    ax.scatter(np.real(poles), np.imag(poles), s=100, marker='x', color='r', label="Poles")

    unit_circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='dashed')
    ax.add_patch(unit_circle)

    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)

    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_title(title)
    ax.legend()
    ax.set_aspect('equal', adjustable='datalim')

    plt.show()

# نمودار اول: X_a(z)
alpha = 0.5  # مقدار آلفا (می‌توان تغییر داد)
zeros_a = [0]  # صفرها
poles_a = [alpha, 1/alpha]  # قطب‌ها
plot_pole_zero(poles_a, zeros_a, "Pole-Zero Plot for X_a(z)")

# نمودار دوم: X_b(z)
N = 10  # مقدار N (می‌توان تغییر داد)
zeros_b = np.exp(2j * np.pi * np.arange(N) / N)  # ریشه‌های واحد
poles_b = [1] * (N - 1)  # قطب‌ها
plot_pole_zero(poles_b, zeros_b, "Pole-Zero Plot for X_b(z)")

# نمودار سوم: X_c(z)
zeros_c = np.exp(2j * np.pi * np.arange(2*N-1) / (2*N-1))  # ریشه‌های دوگانه
poles_c = [1] * (2*N-1)  # قطب‌ها
plot_pole_zero(poles_c, zeros_c, "Pole-Zero Plot for X_c(z)")
```
